```python
# Set the figure size
plt.figure(figsize=(10, 6))

# Create the boxplot using Seaborn with an increased 'whis' value
sns.boxplot(x=df["MINIMUM_PAYMENTS"], whis=3)  # Adjust 'whis' as needed

# Set the title
plt.title("Boxplot with Extended Whiskers")

# Show the plot
plt.show()
```

```python
# fill null values with mean and median values
df["CREDIT_LIMIT"] = df["CREDIT_LIMIT"].fillna(df["CREDIT_LIMIT"].mean())
```

```python
# Replace 
df["MINIMUM_PAYMENTS"] = df["MINIMUM_PAYMENTS"].fillna(df["MINIMUM_PAYMENTS"].median())
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize= (30,45))
for i , col in enumerate(df.columns ) :
    if df[col].dtype != "object":
        ax = plt.subplot(9,2, i+1)
        sns.kdeplot(df[col])

plt.show()
```

```python
col_len=len(df.columns)
```

```python
warnings.filterwarnings('ignore')
plt.figure(figsize=(10,70))
for i in range(0, col_len):
    ax = plt.subplot(17, 1, i+1)
    sns.distplot(df[df.columns[i]], kde_kws= {'color':'gold', 'bw':0.1,'lw':3,'label':'kde'}, hist_kws={'color':'blue'})
    #plt.title(df.columns[i])

plt.show()
```

```python
plt.figure(figsize= (12,12))
sns.heatmap(df.corr(),annot = True)
plt.show()
```
